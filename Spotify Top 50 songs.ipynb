```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv("top50.csv", encoding = "ISO-8859-1", index_col = "Unnamed: 0")
```

```python
data = raw_data.copy()
```

```python
#renaming unnamed column
data.rename(columns = {'Unnamed: 0':'id'}, inplace = True)
```

```python
#Calculating the number of songs of each genre
print(type(data['Genre']))
popular_genre=data.groupby('Genre').size()
print(popular_genre)
genre_list=data['Genre'].values.tolist()
#most popular Genre was POP
```

```python
Genre_list=list(data.groupby("Genre").groups)
print(Genre_list)
print(len(Genre_list))
```

```python
Popularity_list=list(data.groupby("Popularity").groups)
data['Popularity'].describe().mean()
```

```python
from matplotlib import pyplot as plt
data.plot(y='Popularity',x= 'Genre',kind='bar',figsize=(15,2),legend =True,title="Genre V/S Popularity",fontsize=10,stacked=True)
plt.ylabel('Popularity')
plt.xlabel('Genre')
plt.grid(color='black', linestyle='--', linewidth=0.5)
plt.show()
```

```python
data.plot(y='Length.', kind='line',figsize=(10,6))
plt.grid(color='black', linestyle='--', linewidth=0.5)
data.plot(kind='bar',subplots=True, layout=(6,2),figsize=(45,30),color='grey',fontsize=22,legend = True)
```
